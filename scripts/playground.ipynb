## Copy from `uploads/` to `uploads_source/`

```python
path = '/Users/thi/Downloads/math2it/uploads/'
path_copied = '/Users/thi/Downloads/math2it/uploads_source/'
```

```python
for root, _, files in os.walk(path):
    for name in files:
        if not re.match('.*-[0-9]{1,4}x[0-9]{1,4}.*', name):
            # print(name)
            file_full_path = os.path.abspath(os.path.join(root, name))
            rst = shutil.copy(file_full_path, path_copied)
            if (rst):
                print(f'✅ Copied "{name}" to {path_copied}')
```

```python
import pydash

graphql_endpoint = 'http://math2it.local/graphql'
client = GraphQLClient(graphql_endpoint)

final_posts = []

# ==== first posts ====

query = """query postsByCategory {
    posts(first: 100, where: {orderby: {field: DATE, order: DESC}}) {
        nodes {
            databaseId
            title
            content
        }
    }
}"""
result = client.execute(query)
result = json.loads(result)
nodes = result['data']['posts']['nodes']

print(f'#posts: {len(nodes)}')

for node in nodes:
    content = node['content'] if node['content'] else ''
    if '<img src=' in content:
        title = node['title']
        databaseId = node['databaseId']
        final_posts.append({'title': title, 'databaseId': databaseId})

print(f'len(final_posts): {len(final_posts)}')

# ==== last posts ====
query = """query postsByCategory {
    posts(last: 100, where: {orderby: {field: DATE, order: DESC}}) {
        nodes {
            databaseId
            title
            content
        }
    }
}"""
result = client.execute(query)
result = json.loads(result)
nodes = result['data']['posts']['nodes']
print(f'#posts: {len(nodes)}')

for node in nodes:
    content = node['content'] if node['content'] else ''
    if '<img src=' in content:
        title = node['title']
        databaseId = node['databaseId']
        final_posts.append({'title': title, 'databaseId': databaseId})

print(f'len(final_posts): {len(final_posts)}')

# ==== Remove duplicates ====
final_posts = pydash.arrays.uniq_by(final_posts, 'databaseId')
print(f'len(final_posts) after removing duplicates: {len(final_posts)}')

# Save to file
with open('post_with_img.json', 'w', encoding='utf8') as f:
    json.dump(final_posts, f, indent=2, ensure_ascii=False)
```
